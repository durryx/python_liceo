# Problema 2 probabilità
Scelti 2 numeri a e b casuali nell'intervallo $(0, l)$ si possono verificare 2 casi:
1. $a>b$
2. $a<b$

Nel primo caso si avrà un segmento di lunghezza $b$, un segmento di lunghezza $a-b$ ed il segmento rimanente. Poiché per formare un triangolo deve rispettare le disuguaglianze triangolari si ha che:  
$b+(a-b)>l-b-(a-b)$  
$b+l-b-(a-b)>(a-b)$  
$l-b-(a-b)+(a-b)>b$  
Semplificando:  
$a>\dfrac{l}{2}$

$b<\dfrac{l}{2}$

$a<\dfrac{l}{2}+b$  


```python
import numpy as np
import matplotlib.pyplot as plt

l=100 # lughezza del segmento

fig, ax = plt.subplots(figsize=(15,8))
ax.set_xlim(0,100)
ax.set_ylim(0,100)

plt.grid(True)
plt.xlabel("valori del punto b")
plt.ylabel("valori del punto a")
x=np.linspace(0,l/2)
y=x+l/2
x1=np.linspace(l/2,l)
y1=x1-l/2
plt.plot(x,y, color='blue', linewidth=2) # retta n°3 primo caso
plt.plot(x1,y1, color='teal', linewidth=2) # retta n°3 secondo caso
plt.axline((l/2, 0), (l/2, l), color='tan', linewidth=1.5)
plt.axline((0, l/2), (l, l/2), color='lime', linewidth=1.5)
ax.fill_between(x, y, l/2) # riempio l'area caso 1
ax.fill_between(x1, y1, l/2, color='firebrick') # riempio l'area caso 2

plt.show()
```

Calcolando l'area data dalle disequazioni precedenti si ha che $p(triangolo|a>b)=\dfrac{1}{8}$  
Nel secondo caso si ha una situazione analoga, dove però $a$ e $b$ si "scambiano di posto", si avrà quindi un segmento di lunghezza $a$, uno di lunghezza $b-a$ ... ottengo:

$a+(b-a)>l-a-(b-a)$

$a+(l-a-(b-a))>b-a$

$(b-a)+(l-a-(b-a))>a$

Semplificando:

$b>\dfrac{l}{2}$

$a<\dfrac{l}{2}$

$a>b-\dfrac{l}{2}$  

il grafico mostra il risultato di queste disequazioni triangolari in rosso.  
Infine quindi: $p(triangolo)=p(triangolo|a<b)+p(triangolo|a>b)=\dfrac{1}{4}$

```python
import random
import numpy as np
import matplotlib.pyplot as plt

l=100
val=0 # trinagoli validi
i=0
iterazioni=30000

fig, ax = plt.subplots(figsize=(12,10))
ax.set_xlim(0,100)
ax.set_ylim(0,100)
plt.grid(True)
plt.xlabel("valori del punto b")
plt.ylabel("valori del punto a")

while i <= iterazioni:
    a1=a=random.uniform(0.0, l)
    b1=b=random.uniform(0.0, l)     
    if a==b or (0,l) in (a,b):
        continue  # se sono uguali o se uno dei valori è uguale a 100, 0 rieseguo l'iterazione
    elif a>b: # calcolo le lunghezze dei segmenti
        a-=b
        c=l-a-b
    else:
        b-=a
        c=l-a-b
    if a+b>c and a+c>b and b+c>a: # verifico se si può costruire il triangolo
        val+=1
        plt.plot(b1, a1, marker='o', markersize=3, color="r")
    else:
        plt.plot(b1, a1, marker='o', markersize=3, color="k")
    i+=1

plt.show()
print("p(triangolo) = " +str(val/iterazioni)) # verificato il numero
```
