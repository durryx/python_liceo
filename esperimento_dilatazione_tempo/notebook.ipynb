# PSSC - dilatazione del tempo
Lo scopo dell'esperimento è quello di confermare la validità della relatività ristretta, per poi evidenziare come le previsioni della fisica classica non siano in grado di giustificare il fenomeno analizzato.
Utilizzando come orologi i mesoni μ, che sono delle particelle che viaggiano a 0.99c, che decadono in un tempo medio, a riposo, di 2.2 μs (rilevato da esperimenti precedenti), possiamo dimostrare la veridicità della tesi.
## Rilevazione dei muoni
Per rilevare i muoni, utilizziamo uno scintillatore che è uno strumento, che emette un lampo di luce quando una particella carica o radioattiva, nel nostro caso i mesoni, lo attraversa o decade al suo interno.
Per selezionare l'intervallo di velocità dei mesoni che entrano nello scillatore in modo che  
$0.9950c < v_μ < 0.9954c$  
si pone uno spessore di ferro sopra lo scintillatore in modo che le particelle più lente decadano al suo interno e quelle più veloci lo attraversino provocando un solo impulso di luce.

## Misurazioni al Monte Washington
Le misurazioni al Monte Washington riportano che 563 muoni in media decadono all'interno dello scintillatore (quindi si hanno 2 impulsi di luce). Andando al livello del mare (quindi con un dislivello rispetto al monte di 1907 m) quanti muoni decadranno all'interno dello scillatore?

![monte.png](monte.png)

Eleborando una previsione in termini di tempo assoluto: 

Decadimento esponenziale: $ n=n_0e^{(-\tfrac{t}{λ})} ≈ 30 $    

Dove $n_0$ è il numero di muoni misurati (iniziali) ed $n$ il numero di muoni rimasti (che non sono ancora decaduti) in funzione del tempo. Considerando che $ t = \dfrac{∆y}{v_μ} $ dove $ ∆y $ è il dislivello in km, si può calcolare che il tempo di arrivo al livello del mare: $ 6.388 μs < t < 6.386 μs $  
Quindi, secondo questa previsione, si dovrebbero misurare circa 30 muoni.

```python
import math
import numpy as np
import matplotlib.pyplot as plt

n=563
fatt=2.2
t=6.3
n_0=n*math.exp(-t/fatt)

fig, ax = plt.subplots(figsize=(14,10))
ax.spines['right'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
plt.grid(True)


plt.xlabel("Tempo μs")
plt.ylabel("Numero di Muoni")
x=np.linspace(0,15,100)
y=n*np.exp(-x/fatt)
plt.plot(x,y, color='orange', linewidth=2)
plt.plot(t,n_0,'bo')
plt.plot(0,n,'bo')
plt.plot([t,0], [n_0,n_0], '--', color="black", alpha=0.3, lw=1.2)
plt.plot([t,t], [n_0,0], '--', color="black", alpha=0.3, lw=1.2)
plt.annotate("(0 , 563)", (0.2, 560))
plt.annotate("(6.3 , 30)", (6.3, 50))
plt.show()
```

### Ipotesi relativistica
Il tempo si dilata secondo le trasformazioni di Lorentz :

$ γ = \dfrac{1}{\sqrt{1-\dfrac{v_μ^2}{c^2}}} $ &nbsp;&nbsp; $ t' = γ\biggl(t-\dfrac{v_μ{∆y}}{c^2}\biggl)  $  

dove $v_μ$ è la velocità del muone, $ ∆y $ la distanza relativistica percorsa (percepita dalla particella) e $t$ il tempo proprio in secondi. Non è tuttavia possibile applicare l'equazione poiché sia $ t' $, il tempo relativistico misurato dalla terra, sia $ ∆y $, lo spazio visto dalla particella, sono dati mancanti. Di conseguenza l'unica alternativa possibile è di calcolare una approssimazione di $ t' $:  

$ t' ≈ γt $  

Considerando che $ \dfrac{v_μ{∆y}}{c^2} $ ha ordine di grandezza di $ 10^{-6} $ si tratta una approssimazione che non compromette l'esperimento.

## Misurazioni al livello del mare
Al livello del mare si misurano 412 muoni  

$ t=-λlog\biggl(\dfrac{n}{n_0}\biggl) = 0.73 μs $  

La funzione di decadimento è: $ n=n_0e^{(-\tfrac{t}{λ})} ≈ 404 $ mentre dal punto vista della terra: $ n=n_0e^{(-\tfrac{t'}{λγ})} $ dove $ t' $ è 7.38 μs.  
É possibile inoltre calcolare la distanza relativistica $ ∆y' = \dfrac{∆y}{γ} = 188.55 m $

```python
import math
import numpy as np
import matplotlib.pyplot as plt

c=300000
v=0.9951*c

n=563
fatt=2.2
t=0.73     # vedi valore esatto
n_0=404

gamma=1/math.sqrt(1-(v/c)**2)
it=gamma*t

fig, ax = plt.subplots(figsize=(13,9))
ax.spines['right'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
plt.grid(True)
plt.xlabel("Tempo μs")
plt.ylabel("Numero di Muoni")
x=np.linspace(0,15,100)
y=n*np.exp(-x/(fatt*gamma))
plt.plot(x,y, color='lightseagreen', linewidth=2)
plt.plot([it,0], [n_0,n], 'bo')
plt.plot([it,0], [n_0,n_0], '--', color="black", alpha=0.3, lw=1.2)
plt.plot([it,it], [n_0,0], '--', color="black", alpha=0.3, lw=1.2)
plt.annotate("(0 , 563)", [0.2, 560])
plt.annotate("(7.3 , 404)", [7.6, 412])             
plt.show()
```
